```python
from bs4 import BeautifulSoup
import os
import pandas as pd
import pickle
import random
import re
from selenium import webdriver
import time
```

```python
chromedriver = "/Applications/chromedriver" # path to the chromedriver executable
os.environ["webdriver.chrome.driver"] = chromedriver
```

```python
#scraping every apartment listing off of StreetEasy Brooklyn rentals

driver = webdriver.Chrome(chromedriver)
#listings = []

for page_num in range(int(len(listings)/11), 632):
    driver.get(f"https://streeteasy.com/for-rent/brooklyn?page={page_num}")
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    all_urls = soup.find_all('a', class_ = "listingCard-globalLink jsGlobalListingCardLink", attrs = "href")
    for u in all_urls:
        link = u.get("href")
        if link not in listings:
            listings.append(link)
    time.sleep(2.5 + (random.random() * 5.0))
    print("scraped page " + str(page_num))
    
driver.quit()
```
